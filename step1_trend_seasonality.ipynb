```python
import pandas as pd
from fbprophet import Prophet
%matplotlib inline

# read in data and rename columns per fb prophet req's
raw_df = pd.read_csv('ts_traffic.csv')
raw_df['ds'] = pd.to_datetime(raw_df['ds'])
raw_df.sort_values(by=['ds'], inplace=True)
raw_df['y'] = pd.to_datetime(raw_df['Traffic_Volume'])
df = raw_df[['ds', 'y']].copy()
```

```python
# import holidays
holidays = pd.read_csv('./supplemental_data/usholidays.csv')
holidays['ds'] = pd.to_datetime(holidays['ds'])
holidays.sort_values(by=['ds'], inplace=True)
```

```python
m = Prophet(holidays=holidays)
forecast = m.fit(df)
```

```python
forecast = m.predict()
```

```python
fig = m.plot_components(forecast)
```
